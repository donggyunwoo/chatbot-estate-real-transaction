```python
import requests, json
import pandas as pd
from bs4 import BeautifulSoup as bs
import lxml
import yaml

with open('./config.yaml', 'r') as f:
    conf = yaml.safe_load(f)

key = conf['key']


def parsing_to_df(data):
    soup = bs(data, 'lxml-xml')
    if soup.find('resultCode').text!='00':
        print('error : ',soup.find('resultMsg').text)
        return False
    else:
        print('check!!', soup.find('resultMsg').text)
    list_item=soup.findAll('item')
    list_result = []
    for item in list_item:
        temp_dict = dict()
        for info in item.findChildren():
            temp_dict[info.name] = info.text
        list_result.append(temp_dict)
    return pd.DataFrame(list_result)
```

# 아파트

```python
'''아파트 매매 실거래 자료'''
url = 'http://openapi.molit.go.kr:8081/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcAptTrade?'
params ={'serviceKey' : key, 'LAWD_CD' : '11740', 'DEAL_YMD' : '202204' }
response = requests.get(url, params=params)
```
